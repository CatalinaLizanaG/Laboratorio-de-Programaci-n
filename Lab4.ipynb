**Respuestas**

R1. \\
Apache Sparck es un motor unificado diseñado para el procesamiento de datos, y en especial, de grandes magnitudes. Distinguido por utilizar memoria de almacenamiento para cálculos y así optimizar las operaciones.

Las principales ventajas de Apache Spark sobre pandas son que: (1) Pandas realiza una evaluación ansiosa, gestionando mal la memoria; (2) Pandas tiene un mal desempeño cuando se trata de grandes volumenes de datos; (3) Pandas no fue diseñado para soportar computación paralela o distribuida.


R2. \\
Un RDD (Resilient Distributed Dataset) es una colección, cuya distribución inmutable de elementos se organiza en particiones, que constituye el pilar fundamental de Apache Spark. Este modelo permite las operaciones en paralelo de Spark, distribuyendo los datos a través del clúster y logrando de manera eficiente ejecutar distintas operaciones a la vez. Si bien existen structuras más avanzadas para almacenar datos, como los DataFrames, que al ofrecer una interfaz más avanzada se consideran abstracciones de nivel superior sobre los RDD, en Spark siguen siendo fundamentales para entender la arquitectura y la ejecución de tareas.



R3. \\
Las transformaciones generan un nuevo DataFrame con las operaciones deseadas (como filtrar, mapear y agrupar los datos) a partir de uno existente,utilizando la inmutabilidad de los DataFrames de Spark, mientras que las acciones desencadenan la ejecución de todas las transformaciones acumuladas hasta el momento (como show() y count()).

En Spark las transformaciones se organizan y almacenan internamente, luego, solo despues de invocar una acción, se optimiza el plan de ejecución de las transformaciones para ser jecutadas, lo que se denomina `lazy evaluation`.


R4. \\
El particionamiento de datos es muy importante en Spark pues permite optimizar el rendimiento del procesamiento de datos. Esto se logra pues, el particionamiento promueve el paralelismo, lo que facilita a los ejecutores el procesamiento de los datos que están más próximos físicamente, y así reduce el uso de recursos para la transferencia de datos.

De este modo, a cada ejecutor se le asigna una partición específica para trabajar, maximizando la eficiencia del procesamiento y minimizando el tiempo de ejecución.


R5. \\
**Spark Driver:** Encargado de instanciar la `SparkSession`. Es quien se comunica con el gestor del clúster solicitando los recursos que serán utilizados, y tambien el que transforma las operaciones del código en grafos acíclicos dirigidos de tareas, permitiendo su ejecución de manera eficiente.

**Spark Executor:** Se comunica directamente con el programa conductor. Es quien lleva a cabo las tareas asignadas por el conductor en los nodos de trabajo dentro del cluster, procesando datos y devolviendo los resultados al conductor.

R6. \\
Es el optimizador que encuentra la forma más eficiente para ejecutar las operaciones deseadas con los datos. Como se mencionó antes, esto se relaciona al principio de operación de Spark, `lazy evaluation` , el cual permite optimizar el plan de ejecución de las consultas de SQL, encadenando las transformaciones necesarias y a la vez proporcionar tolerancia a fallos.

````python
#. Cree una función **cast_columns** que permita cambiar el tipo de datos de las columnas problemáticas.
#Luego utilice esta función respecto a lo respondido en la pregunta anterior.

def cast_columns(df, cols_types):
    """
    Cambia el tipo de múltiples columnas en un DataFrame de Spark.

    Parámetros:
    - df: DataFrame de Spark.
    - cols_types: Diccionario con nombres de columnas como claves y tipos de datos de Spark como valores.

    Retorna:
    - DataFrame de Spark con tipos de columnas modificados.
    """
    for col_name, new_type in cols_types.items():
        df = df.withColumn(col_name, df[col_name].cast(new_type))
    return df

cols_to_cast = {"Quantity": IntegerType(), "Customer Age": IntegerType(),"Account Age Days": IntegerType(), "Transaction Hour": IntegerType(), "Is an Alien": FloatType()}
df_cast = cast_columns(df, cols_to_cast)
df_cast.printSchema()
````

````python
#1. Agregar una columna llamada "Transaction bp" con el monto total de la transacción en bodoque pesos.
#Se considera que  x  dólares equivalen a  log(48+|x36|)  bodoque pesos.
from pyspark.sql.functions import log, abs, pow
df = df.withColumn( "Transaction bp", log(48+abs(df['Transaction Amount']**36)))

df.show(10)
````

````python
#2. Crear una columna llamada "Transaction Month" con el mes en que se realiza una transacción.
from pyspark.sql.functions import month
df = df.withColumn( "Transaction Month", month(df['Transaction Date']))

df.show(10)
````

````python
#3. Crear la variable Type of purchase según la cantidad de unidades vendidas de acuerdo a las siguientes categorías.
#Compra minorista: 4 productos o menos.
#Compra mayorista: 5 produtos o más.

df = df.withColumn("Type of purchase", when(df['Quantity'] <=4, "Compra minorista").otherwise("Compra mayorista"))
df.show(10)
````

````python
#4. Imprima los registros de compras hechas por alienígenas en el comecio mayorista. Utilice .filter().
df.filter((df['Is An Alien']==1.0) & (df['Type of purchase']=='Compra mayorista')).show(10)
````

````python
#5. Cuente la cantidad de compras realizadas por humanos y la cantidad de compras realizadas por alienígenas. Utilice .groupby().
df.groupBy('Is An Alien').count().show()
````

````python
#6. Muestre una tabla con la recaudación promedio por transacción para cada método de pago, tanto para humanos como alienígenas. Utilice pivot().
df.groupBy("Payment Method").pivot("Is An Alien").mean("Transaction Amount").show()
````

````python
#2. Visualización de Variables Categóricas
# Genere tres gráficos de barras que diferencien entre humanos y aliens.
# Analice y comente cualquier diferencia o similitud observada entre estos dos grupos

import plotly.graph_objects as go
from plotly.subplots import make_subplots

categorical = ['Payment Method', 'Product Category',	'Device Used']
n_cols= 1


fig = make_subplots(
    rows=len(categorical) // n_cols + 1, cols=n_cols, subplot_titles=categorical
)

# para todas las variables categóricas, graficar count de categorias
for idx, col in enumerate(categorical):
    row_idx = idx // n_cols + 1
    col_idx = idx % n_cols + 1

    h = go.Histogram(x= df_pd.loc[df_pd['Is An Alien']==0, col], name='Humanos', marker_color='blue', histnorm="probability")
    a = go.Histogram(x=df_pd.loc[df_pd['Is An Alien']==1, col], name='Aliens', marker_color='green', histnorm="probability")
    fig.add_trace(h, row=row_idx, col=col_idx)
    fig.add_trace(a, row=row_idx, col=col_idx)

    fig.update_xaxes(title_text=f"{col}", row=row_idx, col=col_idx)
    fig.update_yaxes(title_text="Probabilidad", row=row_idx, col=col_idx)

fig.update_layout(
    height=1500,
    title_text="Visualización de Variables Categóricas",
    showlegend=False,
)
fig.show()
````

El equipo de Bodoque e-shop ha solicitado que los datos estén disponibles en una tabla SQL consultable. Además, están interesados en aprovechar las funciones de ventana en SQL para análisis avanzados. Las funciones de ventana permiten realizar cálculos sobre un conjunto de filas que están relacionadas con la fila actual. Por ejemplo, UNBOUNDED PRECEDING se usa para indicar que el rango de la función de ventana comienza desde la primera fila de la partición o del conjunto de resultados, lo cual es útil para calcular sumas acumulativas hasta la fila actual. Las variaciones comunes de este uso incluyen:

- `UNBOUNDED PRECEDING` to `CURRENT ROW`: Calcula desde el inicio de la partición hasta la fila actual.
- `UNBOUNDED PRECEDING` to `UNBOUNDED FOLLOWING`: Cubre todas las filas dentro de la partición.
- `VALUE PRECEDING` to `VALUE FOLLOWING`: Establece un rango específico basado en valores antes y después de la fila actual.

Ejemplo de uso en SQL:

```sql
STAT(COL1_NAME) OVER (PARTITION BY COL2_NAME ORDER BY COL3_NAME ROWS BETWEEN X PRECEDING AND CURRENT ROW)
```

El método que predice la probabilidad es predict_proba

````python
# 3. Crear una función personalizada:
#Elabore una función que prediga la probabilidad de fraude utilizando el último valor de la lista generada por el objeto serializado.
from pyspark.sql.functions import udf
from pyspark.sql.types import FloatType

def custom_function(col1, col2, col3, col4):
  valor = [[col1,col2,col3,col4]]
  prob = object_pkl.predict_proba(valor)
  return float(prob[0][-1])   # prob cometa fraude
````

````python
# 4. Definir la función UDF: Establezca la función UDF con la función personalizada que ha creado.
udf_function = udf(custom_function, FloatType())
````
